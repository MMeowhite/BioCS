## Basic Operation

If you use basic operator `+`, `-`, `*`, `/`, `**` to operate within tensor, the result will be returned by operating with corresponding elements in tensor

```python
x = torch.tensor([1, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])

x + y, x - y, x * y, x / y, x ** y
```

Also, you can use the built-in math function in torch to operate.

```python
torch.exp(x)
```

```python
X = torch.arange(12, dtype=torch.float32).reshape(3, 4)
Y = torch.tensor([
    [2, 1, 4, 3],
    [6, 6, 6, 6],
    [9, 19, 2, 1]
])

X.shape, Y.shape, torch.cat((X, Y), dim = 0), torch.cat((X, Y), dim = 1), torch.cat((X, Y), dim = 0).shape, torch.cat((X, Y), dim = 1).shape
```

Use logic operator `==` to construct logical tensor

```python
X == Y
```

Sum: you will return tensor with only one elements by computing the sum of all elements in the tensor.

```python
X.sum()
```

Despite the different shape, we can use **broadcasting mechanism** to operate tensor by element.

**broadcasting mechanism**:


```python
a = torch.arange(3).reshape(3, 1)
b = torch.arange(2).reshape(1, 2)

a, b, a + b
```

Besides reading, you can modify element by indexing or sliding

```python
X[1, 2] = 9
X
```

```python
X[0:2, :] = 12
X
```

Allocating new memory by some operation.

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print(f"id(Z): {id(Z)}")
Z[:] = X + Y
print(f"id(Z): {id(Z)}")
```

Tips: If you do not reuse the variable, we can also use `X[:] = X + Y` or `X += Y` to reduce memory store.

```python
before = id(X)
x += Y
id(X) == before
```

Numpy transition

```python
A = X.numpy()
B = torch.tensor(A)
type(A), type(B)
```
