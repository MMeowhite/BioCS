# Autograd

Supposing function was determined by $y = 2 x^T x$, we want to make gradiation from column vector

```python
import torch

x = torch.arange(4.0)
x
```

Before calculation of gradient of $y$ about $x$, we need memory to store it.

```python
x.requires_grad_(True) # equvient to x = torch.arange(4.0, requires_grad=True)
x.grad # default value: None
```

Now, let's compute $y$

```python
y = 2 * torch.dot(x, x)
y
```


The gradient of $y$ with respect to each component of $x$ is automatically computed by invoking the backpropagation function.

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

Let's compute another function respect to $x$

```python
# By default, PyTorch accumulates gradients. We need to clear the previous values.
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

In deep learning, our goal is not to compute the differential matrix, but rather the sum of the partial derivatives computed separately for each sample in the batch.

```python
x.grad.zero_()
y = x * x
# equvalent to y.backward(torch.ones(len(x)))
y.sum().backward()
x.grad
```

Move certain computations outside of the recorded computational graph.

```python
x.grad.zero_()
y = x * x
# u = y.detach()
u = y
z = u * x

z.sum().backward()
# x.grad == u
x.grad == 3 * x ** 2
```

```python
def f(a):
  b = a * 2
  while b.norm() < 1000:
    b = b * 2
  if b.sum() > 0:
    c = b
  else:
    c = 100 * b
  return c

a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()

a.grad == d / a
```
